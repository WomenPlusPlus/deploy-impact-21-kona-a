This notebook filters the sub categories that the filtering logic is based on. In the application, when a user selects a Main Category, such as Social Services, it will show a dynamic list of Sub Categories such as Food, Shelter, Water. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###  Create function that splits the list of sub categories into strings. Then count how many times the word appears

This function is useful for finding synonyms that can be grouped together, and for finding misspelled words since I am a terrible speller

```python
def count_frequency(col_name):
  df2 = col_name
  df2_split = df2.str.split(",")
  word_freq = {} 
  for row in df2_split:
    for word in row:
        word_trimmed = word.strip() 
        if (word_trimmed in word_freq): 
            word_freq[word_trimmed] += 1
        else: 
            word_freq[word_trimmed] = 1
  s = pd.Series(word_freq, name='Count')
  s.index.name = "Category"
  s = s.sort_values(ascending=False)
  print(s)
```

## Create a new data frame and force SubCategory to be a string so the list can be split by the function.

```python
df3 = df
df3["SubCategory"] = df3["SubCategory"].astype(str)
```

```python
sub_cat_df2 = df3['SubCategory']
count_frequency(sub_cat_df2)
pd.set_option('display.max_rows', None)
```
